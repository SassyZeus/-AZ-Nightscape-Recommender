```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/drive/My Drive/CIS 509/arizona_businesses_after_22_reviews.csv')  # Replace 'your_dataset.csv' with the path to your dataset

# Group by business_id and find the top 10 most useful reviews for each group
top_10_useful_reviews = data.groupby('business_id').apply(lambda x: x.nlargest(10, 'useful')).reset_index(drop=True)

# Save the dataset to a file
output_file_path = '/content/drive/My Drive/CIS 509/top_10_useful_reviews.csv'  # Specify the file path where you want to save the dataset
top_10_useful_reviews.to_csv(output_file_path, index=False)

print("Dataset saved successfully at:", output_file_path)


```

```python
import pandas as pd

# Load dataset 1
dataset1 = pd.read_csv('/content/drive/My Drive/CIS 509/filtered_businesses_after_22.csv')  # Replace with the path to your dataset 1
# Load dataset 2
dataset2 = pd.read_csv('/content/drive/My Drive/CIS 509/top_10_useful_reviews.csv')  # Replace with the path to your dataset 2

# Selecting only required columns from dataset 1
dataset1_selected_columns = dataset1[['business_id', 'name', 'state', 'city', 'categories','attributes']]

# Selecting only required columns from dataset 2
dataset2_selected_columns = dataset2[['business_id', 'user_id', 'useful', 'text', 'review_id']]

# Merge the datasets on 'business_id'
combined_dataset = pd.merge(dataset1_selected_columns, dataset2_selected_columns, on='business_id', how='inner')

# Specify the file path where you want to save the dataset
output_file_path = '/content/drive/My Drive/CIS 509/dataset_for_topic_modelling2.csv'

# Save the combined dataset to a CSV file
combined_dataset.to_csv(output_file_path, index=False)

print("Dataset saved successfully.")

```
