```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
data_file_path = '/content/drive/MyDrive/CIS 509/dataset_for_topic_modelling2.csv'

# Read the CSV file into a DataFrame
data_df = pd.read_csv(data_file_path)

# Replace NaN values with an empty string
data_df['categories'] = data_df['categories'].fillna('')

# Split categories and count the occurrences of each category
categories = data_df['categories'].str.split(', ', expand=True).stack().value_counts()

# Select top 15 categories
top_categories = categories.head(15)

# Plot the distribution of top 15 categories
plt.figure(figsize=(10, 6))
plt.bar(top_categories.index, top_categories.values, color='blue')
plt.title('Top 15 Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a function to categorize sentiment
def get_sentiment_category(text):
    # Analyze sentiment
    scores = sid.polarity_scores(text)

    # Categorize sentiment based on compound score
    if scores['compound'] >= 0.05:
        return 'Positive'
    elif scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the review text
data_df['sentiment'] = data_df['text'].apply(get_sentiment_category)

# Display the distribution of sentiment categories
sentiment_counts = data_df['sentiment'].value_counts()
print(sentiment_counts)
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of sentiment categories
plt.figure(figsize=(8, 6))
data_df['sentiment'].value_counts().plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Distribution of Sentiment Categories')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.show()
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Load the dataset
file_path = '/content/drive/My Drive/CIS 509/arizona mod.csv'
df = pd.read_csv(file_path)

# Convert the hour strings into datetime objects
def convert_to_datetime(hour_str):
    try:
        start_str, end_str = hour_str.split('-')
        start_time = datetime.strptime(start_str, '%H:%M')
        end_time = datetime.strptime(end_str, '%H:%M')
        # If end time is before start time, add 1 day to end time
        if end_time < start_time:
            end_time += timedelta(days=1)
        return start_time, end_time
    except ValueError:
        return None

# Get the closing time from each time range
def get_closing_time(hours):
    if hours is None:
        return None
    start_time, end_time = hours
    return end_time

# Apply function to get closing time for each day
for day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
    df[day] = df[day].apply(convert_to_datetime)

# Filter out rows with invalid time range format
df.dropna(subset=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], how='any', inplace=True)

# Create new columns with adjusted closing times
for day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
    df[day + '_adjusted'] = df[day].apply(get_closing_time)

# Filter the rows where the closing time is after 19:00 PM
filtered_df = df[df[['Monday_adjusted', 'Tuesday_adjusted', 'Wednesday_adjusted', 'Thursday_adjusted', 'Friday_adjusted', 'Saturday_adjusted', 'Sunday_adjusted']].max(axis=1) > datetime.strptime('19:00', '%H:%M')]

# Save the filtered dataframe to a new CSV file
filtered_file_path = '/content/drive/My Drive/CIS 509/arizona_nightlife_data_review_count_59389_filtered.csv'
filtered_df.to_csv(filtered_file_path, index=False)

print("Filtered dataset saved successfully.")

```

```python
import pandas as pd

# Load the data from the provided CSV file
file_path = '/content/drive/My Drive/CIS 509//arizona mod.csv'
data = pd.read_csv(file_path)
# Display the first few rows of the dataframe to understand its structure
data.head()


# Function to check if a business closes after 19:00 on any day of the week
def closes_after_19(hours_str):
    if hours_str == 'Closed' or pd.isna(hours_str):
        return False
    open_close_times = hours_str.split('-')
    if len(open_close_times) < 2:
        return False
    close_time = open_close_times[1]
    close_hour_minute = close_time.split(':')
    if len(close_hour_minute) < 2:
        return False
    close_hour = int(close_hour_minute[0])
    # Considering businesses that close the next day, e.g., "22:0-2:0"
    if '0' <= close_hour_minute[1] <= '59':
        if close_hour < 12:  # Assuming closing times like "2:0" are AM next day
            return True
    return close_hour >= 19

# Apply the function to each day of the week and check if the business is open
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data['closes_after_19'] = False

for day in days_of_week:
    data['closes_after_19'] |= data[day].apply(closes_after_19)

# Filter data for businesses that are open and close after 19:00 on any day
filtered_data = data[(data['is_open'] == 1) & (data['closes_after_19'])]

# Dropping the 'closes_after_19' column as it was used for filtering
filtered_data.drop(columns=['closes_after_19'], inplace=True)

display(filtered_data)
# Specify the path where you want to save the CSV file
output_file_path = '/content/drive/My Drive/CIS 509/filtered_businesses.csv'

# Save the filtered DataFrame to a new CSV file
filtered_data.to_csv(output_file_path, index=False)

print("Filtered data saved successfully.")
```
